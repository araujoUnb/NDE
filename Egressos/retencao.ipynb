```python
# Carrega os dados do CSV
df = pd.read_csv("dados_eng_eletronica.csv")

# Calcula a taxa de retenção para cada turma
df["Taxa de Retenção (%)"] = (df["Ativos"] / df["Ingressantes"]) * 100

# Exibe a tabela com a nova coluna
df[["Ano-Periodo", "Ingressantes", "Ativos", "Taxa de Retenção (%)"]]
```
